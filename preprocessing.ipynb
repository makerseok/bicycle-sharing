# 데이터 로드

```python
import pandas as pd
```

```python
ride_19_12_20_05 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/19_12_20_05.csv', encoding='cp949')
ride_20_06 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/20_06.csv', encoding='cp949')
ride_20_07 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/20_07.csv', encoding='cp949')
ride_20_08 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/20_08.csv', encoding='cp949')
ride_20_09 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/20_09.csv', encoding='cp949')
ride_20_10 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/20_10.csv', encoding='cp949')
ride_20_11 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/20_11.csv', encoding='cp949')
ride_20_12 = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/20_12.csv', encoding='cp949')
```

```python
station = pd.read_csv('/content/drive/MyDrive/projects/3차-따릉이/data/대여소_목록.csv', encoding='cp949', skiprows=4, names=['대여소번호', '대여소명', '자치구', '상세주소', '위도', '경도', '설치시기', 'LCD', 'QR', '운영방식'])
```

# 데이터 전처리

```python
ride_20_12.shape
```

```python
ride_19_12_20_05.shape
```

```python
# 2019년도 데이터 삭제
ride_2019 = ride_19_12_20_05[ride_19_12_20_05['대여일자'].str.startswith('2019')]
ride_20_01_20_05 = ride_19_12_20_05.drop(ride_2019.index)
```

```python
ride_20_01_20_05
```

```python
# 월별 데이터 하나로 병합
ride = pd.concat([
                  ride_20_01_20_05,
                  ride_20_06,
                  ride_20_07,
                  ride_20_08,
                  ride_20_09,
                  ride_20_10,
                  ride_20_11,
                  ride_20_12,

])
```

```python
# 이용현황에 대여소 번호 추가
ride = ride.merge(station[['대여소번호', '자치구']], on='대여소번호', how='left')
ride
```

```python
# 자치구, 대여시간 별 이용건수
ride = ride[['자치구', '대여시간', '대여일자', '이용건수']].groupby(['자치구', '대여시간', '대여일자']).sum().reset_index()
ride
```

```python
# 파일로 저장
ride.to_csv('/content/drive/MyDrive/projects/3차-따릉이/data/ride_count.csv')
```
